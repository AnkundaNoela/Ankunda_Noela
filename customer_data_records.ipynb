```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#Load the dataset
df = pd.read_csv('customer_data_records.csv')

print(df.head())
```

```python
#features for clustering
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#I choose 3 clusters 
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(x)
```

```python
#cluster labels to dataframe
df['cluster'] = kmeans.labels_
print(df.head())
```

```python
#Visualize the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='cluster', palette='viridis')

plt.title('Customer Clusters based on Income & Spending')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
